## Gradient Descent

```python
##########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################

# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)  # [REQUIRED] Calculate the gradient by evaluating the derivative function at the current point.
    new_x = previous_x - step_multiplier * gradient  # [REQUIRED] Update the current point using the gradient and step_multiplier.

    step_size = abs(new_x - previous_x)

    x_list.append(new_x)
    slope_list.append(df(new_x))  # [REQUIRED] Evaluate the derivative at the new point.

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
```

```python
## TASK-8 : Code the same things as they were done for the first two examples
import numpy as np
import matplotlib.pyplot as plt

# Example cost function (you can replace this with your own)
def cost_function(x):
    return x**2 + 3*x + 2

# Example gradient descent optimization
def gradient_descent(learning_rate, iterations):
    x = 0  # Initial guess
    costs = []  # List to store cost values for each iteration

    for i in range(iterations):
        gradient = 2*x + 3  # Gradient of the cost function
        x -= learning_rate * gradient  # Update x using gradient descent
        cost = cost_function(x)  # Compute the cost at the updated x
        costs.append(cost)  # Append the cost to the list

    return costs

# Set hyperparameters
learning_rate = 0.1
iterations = 20

# Run gradient descent
costs = gradient_descent(learning_rate, iterations)

# Plot reduction in cost for each iteration
plt.plot(range(iterations), costs, marker='o')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title('Reduction in Cost for Each Iteration')
plt.grid(True)
plt.show()
```

## Mean Squared Error: a cost function for regression problems

### $$RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y - \hat{y} \big)^2 $$

```python
# Challenge: Write a python function mse(y, y_hat) returns the MSE? Call the mse(y, y_hat)
# function and print out the MSE for the y_hat calculated above.

def mse(y, y_hat):
    #mse_calc = 1/7 * sum((y - y_hat)**2)
    #mse_calc = (1/y.size) * sum((y - y_hat)**2)
    mse_calc = np.average((y - y_hat)**2, axis=0)
    return mse_calc
```

## $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

## MSE & Gradient Descent

```python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
```
